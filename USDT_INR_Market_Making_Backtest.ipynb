# Basic Market-Making Backtest for Return Calculation
Description: This code provides a simple backtesting framework for a market-making strategy. It tracks the profit or loss for each trade and calculates basic returns without optimizing the trade execution. It helps estimate the performance of a basic market-making strategy by focusing on the daily profit/loss and overall capital.

Key Components:

- Basic market-making strategy simulation
- Calculation of daily profit/loss and overall capital
- Basic ROI calculation
- Visualization of profit/loss performance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def load_data(filepath):
    """
    Load historical price data from a CSV file.
    """
    data = pd.read_csv(filepath, parse_dates=['Date'])
    data.sort_values(by='Date', inplace=True)
    return data

def market_making_backtest(data, initial_capital, spread, trade_fee_percentage):
    """
    Simulate market-making strategy on historical data.
    """
    capital = initial_capital
    units_held = 0
    profit_loss = 0.0
    executed_trades = []
    capital_over_time = []
    daily_profit_loss = []

    for _, row in data.iterrows():
        close_price = row["Close"]
        buy_price = close_price - (spread / 2)
        sell_price = close_price + (spread / 2)
        daily_profit = 0

        # Simulate buy
        if capital >= buy_price:
            executed_buy = buy_price + (trade_fee_percentage / 100) * buy_price
            units_held += 1
            capital -= executed_buy

        # Simulate sell
        if units_held > 0:
            executed_sell = sell_price - (trade_fee_percentage / 100) * sell_price
            units_held -= 1
            capital += executed_sell

            # Calculate profit/loss
            profit_trade = executed_sell - executed_buy
            daily_profit += profit_trade
            profit_loss += profit_trade
            executed_trades.append(profit_trade)

        # Track capital and profit/loss
        daily_profit_loss.append(daily_profit)
        capital_over_time.append(capital + (units_held * close_price))

    final_capital = capital + (units_held * close_price)
    roi = ((final_capital - initial_capital) / initial_capital) * 100

    return final_capital, profit_loss, len(executed_trades), roi, capital_over_time, daily_profit_loss

def plot_results(data, capital_over_time, daily_profit_loss):
    """
    Plot backtest results.
    """
    plt.figure(figsize=(14, 10))

    # Capital Over Time
    plt.subplot(3, 1, 1)
    plt.plot(data["Date"], capital_over_time, label="Capital Over Time", color="blue")
    plt.title("Capital Over Time")
    plt.xlabel("Date")
    plt.ylabel("Capital (INR)")
    plt.legend()

    # Daily Profit/Loss
    plt.subplot(3, 1, 2)
    plt.bar(data["Date"], daily_profit_loss, label="Daily Profit/Loss", color="green", alpha=0.7)
    plt.title("Daily Profit/Loss")
    plt.xlabel("Date")
    plt.ylabel("Profit/Loss (INR)")
    plt.legend()

    # Cumulative Profit/Loss
    cumulative_profit_loss = np.cumsum(daily_profit_loss)
    plt.subplot(3, 1, 3)
    plt.plot(data["Date"], cumulative_profit_loss, label="Cumulative Profit/Loss", color="red")
    plt.title("Cumulative Profit/Loss")
    plt.xlabel("Date")
    plt.ylabel("Cumulative Profit/Loss (INR)")
    plt.legend()

    plt.tight_layout()
    plt.show()

# Parameters
file_path = "Synthetic_USDT_INR_Historical_Data.csv"
initial_capital = 5000.0
spread = 1.0  # Spread in INR
trade_fee_percentage = 0.1  # Maker fee in %

# Run backtest
data = load_data(file_path)
final_capital, profit_loss, total_trades, roi, capital_over_time, daily_profit_loss = market_making_backtest(
    data, initial_capital, spread, trade_fee_percentage
)

# Output results
print(f"Initial Capital: ₹{initial_capital:.2f}")
print(f"Final Capital: ₹{final_capital:.2f}")
print(f"Total Profit/Loss: ₹{profit_loss:.2f}")
print(f"Total Trades Executed: {total_trades}")
print(f"Return on Investment (ROI): {roi:.2f}%")

# Plot results
plot_results(data, capital_over_time, daily_profit_loss)
```

# Market-Making Strategy Backtest with Optimized Trade Execution
Description: This code simulates a market-making strategy using historical price data. It optimizes trade execution by considering both the buy and sell prices with a fixed spread, accounting for trade fees, and tracking capital changes over time. The result provides insights into the overall profitability, return on investment (ROI), and trade execution performance.

Key Components:

- Optimized trade execution (considering spread and fees)
- Capital tracking over time
- ROI calculation and daily profit/loss visualization
- Visualization of capital, profit/loss, and cumulative profit

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

# Load historical data
file_path = "Synthetic_USDT_INR_Historical_Data.csv"
data = pd.read_csv(file_path, parse_dates=["Date"])

# Parameters
initial_capital = 5000.0  # Starting capital in INR

# Function to calculate trade fees
def calculate_fee(price, fee_percentage):
    return (fee_percentage / 100) * price

# Function to simulate market making
def simulate_market_making(data, spread, trade_fee_percentage):
    capital = initial_capital
    units_held = 0
    executed_trades = []
    profit_loss = 0.0

    for _, row in data.iterrows():
        close_price = row["Close"]
        buy_price = close_price - (spread / 2)
        sell_price = close_price + (spread / 2)

        # Attempt buy
        if capital >= buy_price:
            executed_buy = buy_price + calculate_fee(buy_price, trade_fee_percentage)
            units_held += 1
            capital -= executed_buy

        # Attempt sell
        if units_held > 0:
            executed_sell = sell_price - calculate_fee(sell_price, trade_fee_percentage)
            units_held -= 1
            capital += executed_sell

            # Record trade profit
            profit_trade = executed_sell - executed_buy
            profit_loss += profit_trade
            executed_trades.append(profit_trade)

    # Final capital calculation
    final_capital = capital + (units_held * close_price)
    roi = ((final_capital - initial_capital) / initial_capital) * 100

    return final_capital, profit_loss, len(executed_trades), roi

# Parameter optimization
spreads = [0.5, 1.0, 1.5, 2.0]  # Spread options
fees = [0.05, 0.1, 0.15, 0.2]  # Maker fee percentage options
results = []

for spread, fee in product(spreads, fees):
    final_capital, profit_loss, trades, roi = simulate_market_making(data, spread, fee)
    results.append((spread, fee, final_capital, profit_loss, trades, roi))

# Create a results DataFrame for analysis
results_df = pd.DataFrame(results, columns=["Spread", "Fee", "Final Capital", "Profit/Loss", "Trades", "ROI"])
best_result = results_df.loc[results_df["ROI"].idxmax()]

# Visualization
plt.figure(figsize=(12, 6))

# Plot ROI for each combination of spread and fee
for fee in fees:
    subset = results_df[results_df["Fee"] == fee]
    plt.plot(subset["Spread"], subset["ROI"], label=f"Fee: {fee}%")

plt.title("ROI vs Spread for Different Fees")
plt.xlabel("Spread (INR)")
plt.ylabel("ROI (%)")
plt.legend()
plt.grid(True)
plt.show()

# Output best results
print("Optimization Results:")
print(f"Best Spread: {best_result['Spread']} INR")
print(f"Best Fee: {best_result['Fee']}%")
print(f"Final Capital: ₹{best_result['Final Capital']:.2f}")
print(f"Total Profit/Loss: ₹{best_result['Profit/Loss']:.2f}")
print(f"Total Trades Executed: {int(best_result['Trades'])}")
print(f"Return on Investment (ROI): {best_result['ROI']:.2f}%")
```
